```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('./ml-100k/u.data',sep=r'\t',names=['user id','item id', 'rating', 'timestamp'])
filmesId=pd.DataFrame((set(data['item id'])),columns=['Filmes'])

itemData=pd.read_csv('./ml-100k/u.item',sep=r'|',names=['movie_id ', 'movie title' , 'release date' ,' video release date' ,
              'IMDb URL' ,' unknown' , 'Action' , 'Adventure' , 'Animation' ,
             ' Childrens ', 'Comedy' , 'Crime' , 'Documentary' , 'Drama' , 'Fantasy' ,
              'Film-Noir' , 'Horror' , 'Musical' , 'Mystery' , 'Romance' , 'Sci-Fi' ,
              'Thriller' , 'War' , 'Western' ], encoding='latin-1')
userData=pd.read_csv('./ml-100k/u.user',sep='|',names=['user id','age','Genero','Ocupacao','ZipCode'])
```

### Exercicio 1
**Objetivo:** Gerar lista dos 10 filmes mais bem avaliadas usando media simples.

```python
#FORMA 1
def media(vals):
    return sum(vals)/len(vals)
lista=[]
for filme in filmesId['Filmes']:
        x=data[ data['item id']==filme]
        mediaVal=media(x['rating'])
        calculo=len(x['user id'])
        element=[filme,mediaVal,calculo]
        lista.append(element)
melhoresFilmes=pd.DataFrame(lista,columns=['Id_Filme','Media','Quantidade_Avalicao'])
melhores=itemData[itemData['movie_id '].isin([x for x in melhoresFilmes['Id_Filme']])]
melhoresFilmes['Nome']=melhores['movie title'].values
melhoresFilmes.sort_values(by=['Media','Quantidade_Avalicao'],ascending=False,inplace=True)
display(melhoresFilmes.head(10))
```

```python
#FORMA 2
val=data.groupby(['item id']).mean('rating')
val['Quantidade']=data.groupby(['item id']).count()['rating']
val=val.sort_values(by=['rating','Quantidade'],ascending=False).join(itemData.set_index('movie_id ')['movie title']).drop(['user id','timestamp'],inplace=False,axis=1).head(10)
val
```

**ANALISE** 

Podemos ver aqui que utilizando a média simples, percebe-se que os maiores valores são dados por filmes que poucas pessoas avaliaram, pelo fato de não levar em consideração a quantidade de avaliaçoes, logo não necessariamente o filme é bom, ou uma boa analise

### Exercicio 2:

**Objetivo**:  Crie a mesma lista porém utilizando o Damped Means com **k=30**

```python
#FORMA 1
def dampedMeans(vals,k,allVals):
    soma=sum(vals)
    m=media(allVals)
    tamanho=len(vals)
    z=(soma+(k*m))/(tamanho+k)
    return z
lista=[]
for filme in filmesId['Filmes']:
        x=data[ data['item id']==filme]
        dumpVal=dampedMeans(x['rating'],30,data['rating'])
        mediaVal=media(x['rating'])
        calculo=len(x['user id'])
        element=[filme,dumpVal,calculo,mediaVal]
        lista.append(element)
melhoresFilmesDump=pd.DataFrame(lista,columns=['Id_Filme','MediaDump','Quantidade_Avalicao','Media'])
melhoresDump=itemData[itemData['movie_id '].isin([x for x in melhoresFilmesDump['Id_Filme']])]
melhoresFilmesDump['Nome']=melhoresDump['movie title'].values
melhoresFilmesDump.sort_values(by=['MediaDump','Quantidade_Avalicao'],ascending=False,inplace=True)
display(melhoresFilmesDump.head(10))
```

```python
#FORMA 2
y=30*data['rating'].mean()
z=data.groupby(['item id']).count()['rating']
val2=data.groupby(['item id']).apply(lambda x: (x['rating'].sum()+y)/(len(x)+30)).to_frame('MediaDump').join(z.to_frame('Quantidade')).join(itemData.set_index('movie_id ')['movie title']).sort_values(by=['MediaDump','Quantidade'],ascending=False).head(10)
val2
```

**ANALISE**

Como podemos ver essa métrica da um resultado mais satisfatorio, pelo fato  da utilização do damped means que basicamente adiciona recomendações falsas com média global, afim de que tenham mais avaliações para aqueles que tem poucas, isso irá normalizar a media de todos os valores afim de que quem tempoucas avaliações não ganhe com facilidadades dos que tme mais valores.

Por exemplo: no exemplo anterior **Star Kid** teve 3 avlaiações apenas com 5 estrelas, tendo Media 5. Porém sabemos que não tem grande publico que assiste e afirma que o filme é bom. Agora aplicando o damped Means pegamos a media global 30 vezes(como se fosse 30 avaliações novas),que veremos que abaixará a nota dele pois teremos mais avaliações para ele.Quanto a **Star Wars** não altera quase nada pois ja tem muitas avaliações.

### Exercicio 3

**OBJETIVO**:  Considere que o usuário goste de um filme quando a nota que ele atribui a um filme seja
superior ou igual a 4 e não goste caso contrário. Faça uma lista de recomendações dos filmes com
o maior percentual positivo de notas.

```python
#FORMA 1
def percentualPositivas(vals):
    positivas=0
    for val in vals:
          if val>=4:
                positivas=positivas+1
    return (positivas*100)/(len(vals))
lista=[]
for filme in filmesId['Filmes']:
        x=data[ data['item id']==filme]
        percentual=percentualPositivas(x['rating'])
        # mediaVal=media(x['rating'])
        calculo=len(x['user id'])
        element=[filme,percentual,calculo]
        lista.append(element)
melhoresFilmesPercet=pd.DataFrame(lista,columns=['Id_Filme','Percentual','Quantidade_Avalicao'])
melhoresPercet=itemData[itemData['movie_id '].isin([x for x in melhoresFilmesPercet['Id_Filme']])]
melhoresFilmesPercet['Nome']=melhoresPercet['movie title'].values
melhoresFilmesPercet.sort_values(by=['Percentual','Quantidade_Avalicao'],ascending=False,inplace=True)
display(melhoresFilmesPercet.head(10))
```

```python
#FORMA 2
val3=data.groupby('item id').apply(lambda x:  ((
    len(x[x['rating']>=4])*100)/len(x['rating']
    ))).to_frame('Percentual').join(data.groupby('item id').count()['rating'].to_frame('Quantidade')).join(itemData.set_index('movie_id ')['movie title']).sort_values(by=['Percentual','Quantidade'],ascending=False).head(10)
val3
```

**ANALISE**

Aqui a verificação como é basicamente analisada por Filme Bom e filme ruim, e pegar a porcentagem de avaliações boas e rankear os melhores. O problema se dá pois é extramamente injusto com filmes de grande porte que tem diversas avaliações  e a probabilidade desse filme ter o percentual menor que os de menor  porte (que tem menos avaliações). Sendo então não necessariamente são os melhores filmes

### Exercicio 5

**OBJETIVO**:Considerando a mesma condição do exercício anterior, faça uma lista de recomendações
considerando os filmes que possuem mais notas iguais ou superiores a 4 do que notas 1, 2 e 3

```python
#forma 1
def maisBemAvaliados(vals):
    positivas=0
    negativas=0
    for val in vals:
          if val>=4:
                positivas=positivas+1
          else:
                negativas=negativas+1
    return positivas-negativas
lista=[]
for filme in filmesId['Filmes']:
        x=data[ data['item id']==filme]
        maisPositivas=maisBemAvaliados(x['rating'])
        # mediaVal=media(x['rating'])
        calculo=len(x['user id'])
        element=[filme,maisPositivas,calculo]
        lista.append(element)
melhoresFilmesMaisPositv=pd.DataFrame(lista,columns=['Id_Filme','Mais_AvalPositivas','Quantidade_Avalicao'])
melhoresMaisPositiv=itemData[itemData['movie_id '].isin([x for x in melhoresFilmesMaisPositv['Id_Filme']])]
melhoresFilmesMaisPositv['Nome']=melhoresMaisPositiv['movie title'].values
melhoresFilmesMaisPositv.sort_values(by=['Mais_AvalPositivas','Quantidade_Avalicao'],ascending=False,inplace=True)
display(melhoresFilmesMaisPositv.head(10))
```
