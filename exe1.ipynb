```python
import pandas as pd
```

```python
data=pd.read_csv('../ml-100k/u.data',sep=r'\t',names=['user id','item id', 'rating', 'timestamp'])
filmesId=pd.DataFrame((set(data['item id'])),columns=['Filmes'])

itemData=pd.read_csv('../ml-100k/u.item',sep=r'|',names=['movie_id ', 'movie title' , 'release date' ,' video release date' ,
              'IMDb URL' ,' unknown' , 'Action' , 'Adventure' , 'Animation' ,
             ' Childrens ', 'Comedy' , 'Crime' , 'Documentary' , 'Drama' , 'Fantasy' ,
              'Film-Noir' , 'Horror' , 'Musical' , 'Mystery' , 'Romance' , 'Sci-Fi' ,
              'Thriller' , 'War' , 'Western' ], encoding='latin-1')
```

```python
def media(vals):
    return sum(vals)/len(vals)
lista=[]
for filme in filmesId['Filmes']:
        x=data[ data['item id']==filme]
        mediaVal=media(x['rating'])
        calculo=len(x['user id'])
        element=[filme,mediaVal,calculo]
        # if mediaVal==5.0:
        #     lista.append(element)
        if len(lista)<10:
            lista.append(element)
        else:
             valores=[x[1] for x in lista]
             valores=[z for z in lista if min(valores)==z[1]]
             for valor in valores:
                if valor[1]<mediaVal or(valor[1]==mediaVal and valor[2]<calculo):
                    lista.remove(valor)
                    lista.append(element)
                    break
melhoresFilmes=pd.DataFrame(lista,columns=['Id_Filme','Media','Quantidade_Avalicao'])
melhores=itemData[itemData['movie_id '].isin([x for x in melhoresFilmes['Id_Filme']])]
melhoresFilmes['Nome']=melhores['movie title'].values
melhoresFilmes.sort_values(by=['Media','Quantidade_Avalicao'],ascending=False,inplace=True)
display(melhoresFilmes)
```
